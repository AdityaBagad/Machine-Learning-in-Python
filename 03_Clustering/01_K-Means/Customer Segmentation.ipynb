### Importing the required libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
from sklearn.cluster import KMeans 
```

### Reading the Dataset

```python
data = pd.read_csv('Cust_Segmentation.csv')

data.head()
```

### Pre-processing

Address in this dataset is a categorical variable. k-means algorithm isn't directly applicable to categorical variables because Euclidean distance function isn't really meaningful for discrete variables. So, lets drop this feature and run clustering.

```python
data = data.drop('Address', axis=1)
data.head()
```

#### Normalizing

```python
from sklearn.preprocessing import StandardScaler

X = data.values[:,1:]
X = np.nan_to_num(X)
Clus_dataSet = StandardScaler().fit_transform(X)
Clus_dataSet
```

```python
Clus_dataSet.shape
```

### Modeling

```python
clusterNum = 3
k_means = KMeans(init = "k-means++", n_clusters = clusterNum, n_init = 12)
k_means.fit(X)
labels = k_means.labels_
```

```python
labels[:10]
```

### Insights

```python
data["Clus_km"] = labels
data.head()
```

#### Check the centroid values by averaging the features in each cluster.

```python
data.groupby('Clus_km').mean()
```

```python
area = np.pi * ( X[:, 1])**2  
plt.scatter(X[:, 0], X[:, 3], s=area, c=labels.astype(np.float), alpha=0.5)
plt.xlabel('Age', fontsize=18)
plt.ylabel('Income', fontsize=16)

plt.show()
```
