### Import required libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

## Cubic Equation (  $ \ y = a x^3 + b x^2 + c x + d \ $  )

```python
x = np.arange(-5.0, 5.0, 0.1)

a = b = c = 1
#You can adjust the slope and intercept to verify the changes in the graph
y = a*(x**3) + b*(x**2) + c*(x) + 3

y_noise = 20 * np.random.normal(size=x.size)

ydata = y + y_noise

plt.figure(figsize=(8,6))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

## Quadratic Equation (  $ \ Y = X^2 \ $  )

```python
x = np.arange(-5.0, 5.0, 0.1)

y = x**2

y_noise = 2 * np.random.normal(size=x.size)

ydata = y + y_noise

plt.figure(figsize=(8,6))
plt.plot(x, ydata,  'bo')
plt.plot(x,y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

## Exponential (  $ \ Y = a + bc^X \ $  )

```python
X = np.arange(-5.0, 5.0, 0.1)

Y = np.exp(X)

y_noise = 4 * np.random.normal(size=X.size)

ydata = Y + y_noise

plt.figure(figsize=(8,6))
plt.plot(X, ydata,  'bo')
plt.plot(X, Y, 'r') 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

## Logarithmic Equation (  $ y = \log(x)$  )

```python
x = np.arange(1, 50, 0.5)

y = np.log(x)

plt.plot(x,y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.show()
```

## Sigmoid / Logistic Function ( $ Y = \frac{1} {1+ e^{-z}} $ )

```python
z = np.arange(-5.0, 5.0, 0.1)

Y = 1/(1+np.exp(-z))

plt.plot(z, Y) 
plt.ylabel('Dependent Variable')
plt.xlabel('Indepdendent Variable')
plt.grid()
plt.show()
```
